```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import sklearn.linear_model
```

```python
data=pd.read_csv('Oil and Gas 1932-2014.csv')
data
```

```python
x=data['year']
x=np.array(x)
x=x.reshape(-1,1)
```

```python
y=data['oil_price_nom']
y=np.array(y)
```

```python
poly=sklearn.preprocessing.PolynomialFeatures(degree=6)
```

```python
x_poly=poly.fit_transform(x)
```

```python
reg=sklearn.linear_model.LinearRegression()
```

```python
reg.fit(x_poly,y)
```

```python
reg.score(x_poly,y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,reg.predict(x_poly))
```
