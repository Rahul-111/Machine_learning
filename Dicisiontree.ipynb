```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import sklearn.tree
```

```python
data=pd.read_csv('Iris.csv')
```

```python
data['Species'].unique()
```

```python
data.groupby(['Species']).mean()
```

```python
data=data.drop(['Id'],axis=1)
```

```python
y=data['Species']
```

```python
data=data.drop(['Species'],axis=1)
```

```python
data
```

```python
x=data
```

```python
x_train,x_teat,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2)
```

```python
clf_gini=sklearn.tree.DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=3,min_samples_leaf=5)
```

```python
clf_gini.fit(x_train,y_train)
```

```python
clf_gini.score(x_train,y_train)
```

```python
clf=clf_gini.fit(x_train,y_train)
```

```python
sklearn.tree.export_graphviz(clf,out_file='tree.dot')
#convert tree.dot to png
```

```python
data=pd.read_csv('traintitanic.csv')
```

```python
data.groupby(['Survived']).mean()
```

```python
data=data.drop(["PassengerId","Fare","Ticket","Name","Cabin"],axis=1)
```

```python
data=data.join(pd.get_dummies(data['Sex']))
data=data.drop(["Sex"],axis=1)
data=data.join(pd.get_dummies(data['Embarked']))
data=data.drop(["Embarked"],axis=1)
```

```python
data=data.drop(['female'],axis=1)
```

```python
data
```

```python
data=data.dropna()
y=data["Survived"]
data=data.drop(['Survived'],axis=1)
```

```python
x=data
x
```

```python
x_train,x_teat,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2)
clf_gini=sklearn.tree.DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=3,min_samples_leaf=5)
```

```python
clf_gini.fit(x_train,y_train)
```

```python
clf_gini.score(x_train,y_train)
```

```python
clf=clf_gini.fit(x_train,y_train)
sklearn.tree.export_graphviz(clf,out_file='treetitanic.dot')
```
