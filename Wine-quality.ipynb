```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('winequality-data.csv')
```

```python
data
```

```python
fixedacidity1=[]
volatileacidity1=[]
fixedacidity2=[]
volatileacidity2=[]
fixedacidity3=[]
volatileacidity3=[]

for i in range(len(data)):
    if data['quality'][i] in range(0,5):
        fixedacidity1.append(data['fixed.acidity'][i])
        volatileacidity1.append(data['volatile.acidity'][i])
    elif data['quality'][i] in range(5,8):
        fixedacidity2.append(data['fixed.acidity'][i])
        volatileacidity2.append(data['volatile.acidity'][i])
    else:
        fixedacidity3.append(data['fixed.acidity'][i])
        volatileacidity3.append(data['volatile.acidity'][i])
```

```python
plt.scatter(fixedacidity1,volatileacidity1,label='Poor')
plt.scatter(fixedacidity2,volatileacidity2,label='Average')
plt.scatter(fixedacidity3,volatileacidity3,label='Premium')
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
```

```python
%matplotlib notebook
```

```python
plt.legend()
```

```python
citricacid1=[]
residualsugar1=[]
citricacid2=[]
residualsugar2=[]
citricacid3=[]
residualsugar3=[]

for i in range(len(data)):
    if data['quality'][i] in range(0,5):
        citricacid1.append(data['citric.acid'][i])
        residualsugar1.append(data['residual.sugar'][i])
    elif data['quality'][i] in range(5,8):
        citricacid2.append(data['citric.acid'][i])
        residualsugar2.append(data['residual.sugar'][i])
    else:
        citricacid3.append(data['citric.acid'][i])
        residualsugar3.append(data['residual.sugar'][i])
```

```python
plt.scatter(citricacid1,residualsugar1,label='Poor')
plt.scatter(citricacid2,residualsugar2,label='Average')
plt.scatter(citricacid3,residualsugar3,label='Premium')
plt.xlabel('Citric Acid')
plt.ylabel('Residual Sugar')
```

```python
plt.legend()
```
