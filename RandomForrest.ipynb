```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import sklearn.datasets
import sklearn.ensemble
```

```python
data=pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
y=data['Species']
```

```python
data=data.drop(['Id','Species'],axis=1)
```

```python
data.head()
```

```python
x=data
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2)
```

```python
clf=sklearn.ensemble.RandomForestClassifier(random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
clf.feature_importances_
```

```python
data=pd.read_csv('traintitanic.csv')
```

```python
data
```

```python
data=data.drop(["PassengerId","Fare","Ticket","Name","Cabin"],axis=1)
```

```python
data=data.join(pd.get_dummies(data['Sex']))
```

```python
data=data.drop(["Sex"],axis=1)
data=data.join(pd.get_dummies(data['Embarked']))
data=data.drop(["Embarked"],axis=1)
```

```python
data=data.drop(['female'],axis=1)
```

```python
data.shape
```

```python
data=data.dropna()
y=data["Survived"]
x=data.drop(['Survived'],axis=1)
```

```python
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2)
```

```python
clf=sklearn.ensemble.RandomForestClassifier(random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
x
```

```python
clf.feature_importances_ #Age is most import
```
