```python
data=pd.read_csv('Oil and Gas 1932-2014.csv')
```

```python
data
```

```python
pro=[]
price=[]
for i in range(len(data)):
    if data['gas_prod55_14'][i] >=0:
        pro.append(data['gas_prod55_14'][i])
        price.append(data['gas_price_nom'][i])
```

```python
plt.scatter(pro,price)
```

```python
reg=sklearn.linear_model.LinearRegression()
```

```python
x=np.array(pro)
y=np.array(price)
x=x.reshape(-1,1)
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```
