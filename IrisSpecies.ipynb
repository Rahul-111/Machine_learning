```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('iris.csv')
```

```python
data
```

```python
PetalLengthCm1=[]
PetalWidthCm1=[]

PetalLengthCm2=[]
PetalWidthCm2=[]

PetalLengthCm3=[]
PetalWidthCm3=[]

for i in range(len(data)):
    if data['Species'][i]=='Iris-setosa':
        PetalLengthCm1.append(data['PetalLengthCm'][i])
        PetalWidthCm1.append(data['PetalWidthCm'][i])
    if data['Species'][i]=='Iris-versicolor':
        PetalLengthCm2.append(data['PetalLengthCm'][i])
        PetalWidthCm2.append(data['PetalWidthCm'][i])
    if data['Species'][i]=='Iris-virginica':
        PetalLengthCm3.append(data['PetalLengthCm'][i])
        PetalWidthCm3.append(data['PetalWidthCm'][i])
```

```python
plt.scatter(PetalLengthCm1,PetalWidthCm1,label='Iris-setosa')
plt.scatter(PetalLengthCm2,PetalWidthCm2,label='Iris-versicolor')
plt.scatter(PetalLengthCm3,PetalWidthCm3,label='Iris-virginica')
plt.title('Scatter Plot')
plt.xlabel('Patel Length')
plt.ylabel('Patel Width')
```
