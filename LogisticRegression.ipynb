```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import sklearn.linear_model
```

```python
data=pd.read_csv('traintitanic.csv')
```

```python
data=data.drop(["PassengerId","Fare","Ticket","Name","Cabin"],axis=1)
```

```python
data=data.join(pd.get_dummies(data['Sex']))
```

```python
data=data.drop(["Sex"],axis=1)
data=data.join(pd.get_dummies(data['Embarked']))
data=data.drop(["Embarked"],axis=1)
```

```python
data=data.drop(['female'],axis=1)
data
```

```python
data=data.dropna()
y=data["Survived"]
x=data.drop(['Survived'],axis=1)
```

```python
reg=sklearn.linear_model.LogisticRegression()
```

```python
reg.fit(x,y)
reg.score(x,y)
```
